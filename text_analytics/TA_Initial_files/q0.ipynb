```python
import os
import json
import gzip
import pandas as pd
from urllib.request import urlopen
```

# Load first dataset- Musical instruments

```python
file1=r"C:\Users\User\Downloads\Musical_Instruments_5.json.gz"


data = []
with gzip.open(file1) as f:
    for l in f:
        data.append(json.loads(l.strip()))
    
# total length of list, this number equals total number of products
print(len(data))

# first row of the list
print(data[0])
```

```python
# convert list into pandas dataframe

df = pd.DataFrame.from_dict(data)

print(len(df))
```

```python
df
```

```python
df.describe()
```

```python
df['Category']='Musical_instruments'
```

```python
musical_data=df[['overall','reviewText']]
musical_data.shape
```

# Load second dataset- Cellphones and Accessories

```python
file2=r"C:\Users\User\Downloads\Cell_Phones_and_Accessories_5 (1).json.gz"
data1 = []
with gzip.open(file2) as f:
    for l in f:
        data1.append(json.loads(l.strip()))
    
# total length of list, this number equals total number of products
print(len(data1))

# first row of the list
print(data1[0])
```

```python
# convert list into pandas dataframe

df1 = pd.DataFrame.from_dict(data1)

print(len(df1))
```

```python
df1.head(2)
```

```python
df1['Category']='cellphones'
```

```python
cells_data=df1[['overall','reviewText']]
cells_data.shape
```

# Combine cellphones and musical instruments data

```python
combined_data=pd.concat([musical_data,cells_data], ignore_index=True)
combined_data.head(2)
```
