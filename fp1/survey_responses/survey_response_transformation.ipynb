```python
import pandas as pd
```

```python
# Reading survey responses
excel=r"food_recommendation_survey_results.xlsx"
sheet_name='Survey Response'
base_data = pd.read_excel(excel,sheet_name=sheet_name)
base_data.head()
```

```python
base_data.drop(['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2'], axis=1, inplace=True)
```

```python
base_data.columns = base_data.iloc[0].values
```

```python
base_data.drop(0, inplace=True)
```

```python
base_data.head()
```

```python
# Reading Food details
food_details = pd.read_excel(r"C:\Users\kanis\OneDrive - Indian School of Business\Desktop\ISB\github_codes\ampba_24_group18_term2\fp1\webapp_sample_datasets\food_details.xlsx")
food_details.head()
```

```python
food_details_1 = food_details.copy(deep=True)
food_details_1.columns = [x+'_1' for x in food_details_1.columns]
joined_data = base_data.merge(food_details_1[['Name_1','Food_ID_1']], left_on='Food Item_1', right_on='Name_1', how='left')
```

```python
joined_data.head()
```

```python
# Checking if any food items didn't match
joined_data[joined_data['Name_1'].isna()]['Food Item_1'].value_counts()
```

```python
food_details_2 = food_details.copy(deep=True)
food_details_2.columns = [x+'_2' for x in food_details_2.columns]
joined_data = joined_data.merge(food_details_2[['Name_2','Food_ID_2']], left_on='Food Item_2', right_on='Name_2', how='left')
```

```python
food_details_3 = food_details.copy(deep=True)
food_details_3.columns = [x+'_3' for x in food_details_3.columns]
joined_data = joined_data.merge(food_details_3[['Name_3','Food_ID_3']], left_on='Food Item_3', right_on='Name_3', how='left')
```

```python
joined_data['user_id'] = range(1, len(joined_data['Mood']) + 1)
joined_data
```

```python
rating_data_raw = joined_data[['user_id', 'Food_ID_1', 'Rating_1', 'Food_ID_2', 'Rating_2', 'Food_ID_3', 'Rating_3']]
rating_data_raw
```

```python
rating_data = pd.melt(rating_data_raw, id_vars=['user_id'], 
                  value_vars=['Food_ID_1', 'Rating_1', 'Food_ID_2', 'Rating_2', 'Food_ID_3', 'Rating_3'],
                  var_name='food_rating', value_name='value')
rating_data
```

```python
# Separate food IDs and ratings
food_df = rating_data[rating_data['food_rating'].str.contains('Food_ID')].rename(columns={'Value': 'Food_ID'}).drop('food_rating', axis=1)
rating_df = rating_data[rating_data['food_rating'].str.contains('Rating')].rename(columns={'Value': 'Rating'}).drop('food_rating', axis=1)

# Merge them back to get the final DataFrame
final_df = pd.concat([food_df.reset_index(drop=True), rating_df.reset_index(drop=True)], axis=1)
final_df
```

```python
final_df = final_df.T.drop_duplicates().T
```

```python
final_df.columns = ['user_id', 'food_id', 'rating']
```

```python
final_df[final_df['user_id'] == 1]
```

```python
final_df.to_csv('ratings_data.csv')
```

```python
# User Attributes
user_attributes = joined_data[['user_id', 'Mood', 'Cuisine']]
user_attributes
```

```python
combinations = pd.DataFrame(columns=['user_id', 'Mood', 'Cuisine'])
i = 2000

uniq_mood = user_attributes['Mood'].unique()
uniq_cuisine = user_attributes['Cuisine'].unique()

for mood in uniq_mood:
    for cuisine in uniq_cuisine:
        combinations = combinations._append({'user_id': i, 'Mood': mood.strip(), 'Cuisine': cuisine.strip()}, ignore_index=True)
        i+=1
combinations.head()
```

```python
all_user_attributes = pd.concat([user_attributes, combinations])
all_user_attributes.head()
```
